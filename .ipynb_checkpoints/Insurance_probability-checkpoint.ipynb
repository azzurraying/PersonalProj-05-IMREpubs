# Background of problem

Couples sign up for the insurance and pay a premium of 199. If they get married after 3 years, but not later than 13 years, they get a 10000-dollar diamond for free.

- premium = cost of signing premium
- reward = cost of prize
- n = number of couples who signed up
- Y = number of couples who made it (got married after 3 years have passed)
- epsilon = profit margin = net income / net sales = (premium x n - reward x Y) / (premium x n)

# Import libraries

```python
import scipy.stats
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy import linspace
import pandas as pd
%matplotlib inline
```

# Define functions

```python
def getProfitProbability(premium = 199, reward = 10000, epsilon = 0.05, n = 1000, p_marry = 0.5):
    Y = premium * n * (1-epsilon) / reward
    
    mean_marry = float(p_marry) * n
    var_marry = float(p_marry) * (1-p_marry) * n
    
    Y_bar = (Y - mean_marry) / var_marry**0.5 # The z score
    return scipy.stats.norm(0, 1).cdf(Y_bar)
```

```python
def getLogProfitProbability(premium = 199, reward = 10000, epsilon = 0.05, n = 1000, p_marry = 0.5):
    Y = premium * n * (1-epsilon) / reward
    
    mean_marry = float(p_marry) * n
    var_marry = float(p_marry) * (1-p_marry) * n
    
    Y_bar = (Y - mean_marry) / var_marry**0.5 # The z score
    if scipy.stats.norm(0, 1).cdf(Y_bar) != 0:
        return math.log(scipy.stats.norm(0, 1).cdf(Y_bar), 10)
    else:
        return -1000
    
```

# Preliminary analysis

```python
premium = 199
reward = 10000
n = 10000
epsilon = 0.05
Y = premium * n * (1-epsilon) / reward

print Y
```

The company could hit a 5% profit margin (http://www.investopedia.com/ask/answers/052515/what-usual-profit-margin-company-insurance-sector.asp) if less than 190 couples, out of 10K, get married within the year.

What is the probability that this happens? I.e., want:

P(Y < 190)

To do this analysis, first treat the probability of each couple getting married as a binomial distribution. I.e.

p_marry = probability of success (gets married) for each trial (couple)

n = number of trials

Given the sample size, the mean and variance for the number of couples that get married are:

mean_marry = p_marry x n

var_marry = p_marry(1-p_marry) x n

When n is large, the binomial distribution is approximately normal. The mean and var can be taken to be those of the normal distribution:

```python
p_marry = 0.5
mean_marry = float(p_marry) * n
var_marry = float(p_marry) * (1-p_marry) * n

print mean_marry
print var_marry
```

We need to normalize the Y variable (in order to compute the probability under a normal curve). I.e., if we want

P(Y < 190)

we want to compute

P(Y_bar < ?)

Where Y_bar = (Y - mean_marry) / sqrt(var_marry)

```python
Y_bar = (Y - mean_marry) / var_marry**0.5
print Y_bar
```

So we want to compute

P(Y_bar < -96)

Without actually computing the probability, we can tell it's vanishingly small, because that's 96 standard deviations below the mean! But let's just for comprehensiveness' sake calculate the actual value:

```python
print scipy.stats.norm(0, 1).cdf(Y_bar)
```

# Optimize parameters

Now let's run a cross-grid to get a feeling for how parameters premium, reward, n and p_marry affect the chance of earning 5% profit.

## Premium

```python
for i in np.arange(0, 2000, 200):
    print getProfitProbability(premium = i, reward = 10000, epsilon = 0.05, n = 1000, p_marry = 0.5)
```

##### CONCLUSION #1: Adjusting the premium doesn't impact the insurance company's profit. It's better to let it stay at the usual rate.

## Reward

```python
for i in np.arange(1, 10000, 1000):
    print getProfitProbability(premium = 199, reward = i, epsilon = 0.05, n = 1000, p_marry = 0.5)
```

```python
for i in np.arange(1, 1000, 100):
    print getProfitProbability(premium = 199, reward = i, epsilon = 0.05, n = 1000, p_marry = 0.5)
```

##### CONCLUSION #2: To have even a chance at earning a 5% profit margin, the reward needs to be below 500 dollars. This doesn't make sense as nobody will pay 199 and wait 3 years to get 500.

## Number of couples who sign up

```python
for i in np.arange(1, 10000, 500):
    print getProfitProbability(premium = 199, reward = 10000, epsilon = 0.05, n = i, p_marry = 0.5)
```

##### CONCLUSION #3: It turns out smaller number of participants heightens the chances of the insurance company earning a profit. Any number greater than 100 decreases the chances precipitously.

This doesn't make a lot of sense since one expect the number of couples who sign up to be in the 1000s if not 10000s. It makes sense, however, if the insurance company limits signups, or if the probability of marriage is way lower than by chance.

## Probability of marriage in three years 

```python
for i in np.arange(0, 0.1, 0.01):
    print getProfitProbability(premium = 199, reward = 10000, epsilon = 0.05, n = 1000, p_marry = i)
```

##### CONCLUSION #4: The likelihood of couples getting married turns out to be the most sensitive parameter in determining whether the insurance company earns any money. The company needs to count on the couples breaking up within 3 years: the company has a 40% chance of getting 5% profit margin if the probability of marriage for each couple after 3 years (and before 13 years) is 0.02. This falls to 2% if the probability of marriage is 0.03. 

## Visualize how n and p_marry impact profit together

```python
for i in np.arange(100, 1000, 200):
    for j in np.arange(0, 0.05, 0.01):
        print 'n =', i, '; p_marry =', j
        print getProfitProbability(premium = 199, reward = 10000, epsilon = 0.05, n = i, p_marry = j)
    print '\n'
```

```python
for i in np.arange(1000, 10000, 1000):
    for j in np.arange(0, 0.03, 0.005):
        print 'n =', i, '; p_marry =', j
        print getProfitProbability(premium = 199, reward = 10000, epsilon = 0.05, n = i, p_marry = j)
    print '\n'
```

##### CONCLUSION #5: The chance of marriage needs to be no higher than 2% for the insurance company to earn a profit.  Also, under such circumstances, greater signup numbers secures the chance of profit. Therefore it's advantageous to sign up huge numbers of couples as long as the prevalent marriage rate is low.

While a marriage rate of below 2 percent sounds very low, we don't have enough data to validate this belief of the insurance company. On the other hand, if a clause is added such that not only a marriage needs to take place between Year 3 and 13, and a divorce must not take place between that time period, the low probability of success would be way more believable.

```python
plt.rcParams.update(pd.tools.plotting.mpl_stylesheet)
colors = pd.tools.plotting._get_standard_colors(len(np.arange(1000, 10000, 2000)), color_type='random')
# 'colors' is a list of 5 lists, each sub-list contains 3 floats, representing r, b, g values.

fig, ax = plt.subplots(figsize = (10, 10))
ax.set_color_cycle(colors)
ax.margins(0.05)

plt.xlim(0, 0.03)
plt.ylim(0, 1.05)

plt.yticks(np.arange(0, 1, 0.1), fontsize=14) # (math intervals, label text, fontsize)
plt.xticks(fontsize=14)    

plt.xlabel('Probability of marriage between years 4 and 13')
plt.ylabel('Probability of insurance company earning 5% profit')

for n_signup in np.arange(1000, 10001, 2000):
    x = np.arange(0, 0.035, 0.001)
    y = [getProfitProbability(n=n_signup, p_marry=prob) for prob in np.arange(0, 0.035, 0.001)]
    ax.plot(x, y, linestyle='solid', ms=10, label=str(n_signup)+' signups', linewidth=3)

ax.legend(numpoints=1, loc='upper left')
plt.title("Probability of profitting from selling marriage insurance", fontsize= 20)    

plt.savefig("marriage_insurance_probability_color1.png", bbox_inches="tight")
```
